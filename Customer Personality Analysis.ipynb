```python
df=pd.read_csv('D:\\marketing_campaign.csv')
```

```python
#Total Purchase 
df['Total Purchase']=df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
```

```python
# purchases w.r.t marital status
p_by_ms=df.groupby('Marital_Status')['Total Purchase'].sum().reset_index()
p_by_ms
```

```python
#Visualizing 
plt.figure(figsize=(10,6))
sns.barplot(data=p_by_ms,x='Marital_Status',y='Total Purchase')
```

# Observation
1.Largerst number of purchase made by 'Married' people followed by 'Together' and 'Single'.

```python
#EDucation vs Purcahses
p_by_ed=df.groupby('Education')['Total Purchase'].sum().reset_index()
p_by_ed
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=p_by_ed,x='Education',y='Total Purchase')
```

# Observation
2.Largest purchase done by Graduates.

```python
grd
```

## Trying explore data w.r.t Date

```python
#convert object to datetime
df['Date_enrolled']= pd.to_datetime(df['Dt_Customer'])
```

```python
df.drop('Dt_Customer',inplace=True,axis=1)
```

```python
df['Date']= df['Date_enrolled'].dt.to_period('M')
```

```python
df['Year']=df['Date'].dt.year
```

```python
df
```

```python
# Year vs Total purchase
dp=df.groupby('Year')['Total Purchase'].sum().reset_index().sort_values('Total Purchase',ascending=False).head(10)

plt.figure(figsize=(25,6))
sns.barplot(data=dp,x= 'Year',y= 'Total Purchase')
```

```python
# want to see in which year most customer accepted offer
df["total_campaign_purchase"]= df['AcceptedCmp3'] + df['AcceptedCmp4'] +df['AcceptedCmp5'] + df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['Response']
```

```python
total_amt_purchase_year= df.groupby('Year')["total_campaign_purchase"].sum().reset_index()
total_amt_purchase_year
```

# Observation
